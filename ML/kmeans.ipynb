```python
features = [
    "QUANTITYORDERED",
    "PRICEEACH",
    "SALES",
    "QTR_ID",
    "MONTH_ID",
    "YEAR_ID",
    "PRODUCTLINE",
    "STATUS",
    "MSRP",
    "COUNTRY",
    "DEALSIZE",
]

df = df.loc[:, features]
df.isnull().sum().sort_values(ascending=False)
```

```python
status_mapping = {
    "Shipped": 0,
    "Resolved": 1,
    "Cancelled": 2,
    "On Hold": 3,
    "Disputed": 4,
    "In Process": 5,
}
df["STATUS"] = df["STATUS"].map(status_mapping)
df
```

```python
dealsize_mapping = {"Small": 0, "Medium": 1, "Large": 2}
df["DEALSIZE"] = df["DEALSIZE"].map(dealsize_mapping)
df
```

```python
cat_cols = df.select_dtypes(include=["object"]).columns
cat_cols
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
processed = ohe.fit_transform(df[cat_cols])
processed_df = pd.DataFrame(processed, columns=ohe.get_feature_names_out())

df = pd.concat([df.drop(columns=cat_cols, axis=1), processed_df], axis=1)
df
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_df = pd.DataFrame(
    scaler.fit_transform(df), columns=scaler.get_feature_names_out()
)
scaled_df
```

```python
from sklearn.decomposition import PCA

decomposer = PCA(n_components=2)
pca_df = pd.DataFrame(decomposer.fit_transform(scaled_df), columns=["PC1", "PC2"])
pca_df


print("Explained variance ratio:", decomposer.explained_variance_ratio_)
print("Total variance retained:", np.sum(decomposer.explained_variance_ratio_))
```

```python
custom_colors1 = ["#01FCD5", "#FA9189", "#AB80FE"]
plt.figure(figsize=(20, 10))
sns.scatterplot(
    data=pca_df,
    x="PC1",
    y="PC2",
    markers=[".", "*", "o"],
    hue=df["DEALSIZE"],
    edgecolor="black",
    palette=custom_colors1,
)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertias = []
silhouette_scores = []
ks = list(range(2, 15))
for k in ks:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = kmeans.fit_predict(pca_df)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(pca_df, labels))
```

```python
index = silhouette_scores.index(max(silhouette_scores))
k_selected = ks[index]
inertia_selected = inertias[index]
silhouette_score_selected = silhouette_scores[index]
print(k_selected, inertia_selected, silhouette_score_selected)
```

```python
plt.figure(figsize=(20, 10))
plt.plot(ks, inertias, "o-")
plt.plot(3, inertia_selected, "*", markersize=10)
plt.title("Elbow Methods (K vs Inertia)")
plt.xticks(range(2, 15, 1))
plt.xlabel("K")
plt.ylabel("Inertia")
```

```python
plt.figure(figsize=(20, 10))
plt.plot(ks, silhouette_scores, "o-")
plt.plot(5, silhouette_scores[3], "*", markersize=10)
plt.title("K vs Silhouette Scores")
plt.xticks(range(2, 15, 1))
plt.xlabel("K")
plt.ylabel("Silhouette Score")
```

```python
kmeans_final = KMeans(n_clusters=k_selected, n_init=10, random_state=42)
labels_final = kmeans_final.fit_predict(scaled_df)
```

```python
pca_df["cluster"] = labels_final
pca_df
```

```python
centers_scaled = kmeans_final.cluster_centers_
centers_scaled_df = pd.DataFrame(centers_scaled, columns=scaled_df.columns)
centers_pca = decomposer.transform(centers_scaled_df)
```

```python
custom_colors2 = ["#FA9189", "#FCFDA4", "#01FCD5", "#0CE3FE", "#AB80FE"]
plt.figure(figsize=(20, 10))
sns.scatterplot(
    data=pca_df,
    x="PC1",
    y="PC2",
    hue="cluster",
    edgecolor="black",
    palette=custom_colors2,
)
```

```python
final_silohuette_score = silhouette_score(pca_df, labels_final)
print(final_silohuette_score)
```
